```python
sns.displot(data=mpg ,x='cylinders',kind='ecdf') # ecdf shows proportion density
```

we can shows relationship between 2 variables by plotting them on x and y axis.such plot known as bivariate(2 variable plot). show cylinder on x axis and mileage on y axis

```python
# to draw distribution plot
# load the mpg dataset into a datframe
mpg=sns.load_dataset('mpg')
```

```python
## distribution plot cylinder on x axis and miles per gallon on y axis
sns.displot(data=mpg,x="cylinders",y="mpg")
plt.xlabel('no of cylinders')
plt.ylabel('miles per gallon')
plt.show()
```

```python
sns.displot(data=mpg,x='cylinders',y='mpg',kind='kde')
```

**BOX PLOT* shows the distribution of data in form of boxes.the shows quartile of data. quartile means 1/4th or quartile. value divides the series in 4 equal parts. shown as q1,q2,q3,q4
the box starts with Q1 ans end with Q3. difference between Q3-Q1 =IQR


a box plot  s( box and whisker plot) shows how numerical values are spread out. distribution and outliers
help u see
>minimun

```python
sns.boxplot(x='origin',y='acceleration', data=mpg)
plt.show()
```

**Scatter plot**
display a groupm of dots (points ) scattered all over xy plane. represents relation between 2 variable.

```python
sns.scatterplot(x='horsepower',y='acceleration',data=mpg,hue='origin')
plt.show()
```

**Positive relationship between x and y:**
if x horsepower increases, y acceleration increases
if x horsepower decreases, y acceleration decreses

**Negative relationship between x and y:**
if x horsepower increases, y acceleration decreases
if x horsepower decreases, y acceleration incraeses

 in the above graph horsepower inc the acceleration dec this hows negative relationship

**Joint plot**

 is the combination of scatter plot and histogram.

```python
sns.jointplot(x='horsepower',y='acceleration',data=mpg)
plt.show()
```

**Line plot **
lineplots connects all the dots with a straight line. helpful to know when values are becoming high and when they are low

```python
sns.lineplot(x='horsepower',y='acceleration',data=mpg,hue='origin')
```
