```python
#basic data cleaning operation without jump into statistical or visualization techmique first basic
# summarize the number of unique values for each column using numpy
import pandas as pd
import numpy as np

```

```python
# load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/oil-spill.csv')
# summarize the number of unique values in each column
df
```

```python
df.head(5)
```

```python
df.columns
```

```python
df.nunique()    # identifying unique values for each columns
```

```python
df.drop(columns='0', axis =1, inplace=True)
```

```python
df.shape # 0 column got deleted
```

```python
df.isnull().sum()
```

```python
#empty list to store selected column namespace
low_unique_cols = []


#loop through each column
for i in df.columns:
     np.unique_counts = df[i].nunique()
     if np.unique_counts <10:
          low_unique_cols.append(i)

print("columns with <10 unique values")
print(low_unique_cols)
```

```python
# other method to find low_unique _cols
####
#low_unique_cols = [col for col in df.columns if df[col].nunique() < 10]
```

```python
## check whether the low_unique_column is useful for prediction or not if not de;lete it

df.drop(low_unique_cols, axis=1, inplace = True)
```
