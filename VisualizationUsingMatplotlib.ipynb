```python
import matplotlib.pyplot as plt  #for drawing graph
import pandas as pd # for creating dataframe
```

```python
empdata={"empid":[1001,1002,1003,1004,1005,1006],"ename":["kalai","jasmine","deepi","priya","gayathri","lakshmi"],"sal":[10000,12000,13000,14000,15000,16000],
}
```

**Bar graph**

```python
x=df["empid"]
y=df["sal"]
```

```python
plt.bar(x,y,label="employee data", color="red")
plt.xlabel("employee id")
plt.ylabel("employee salary")
plt.title("ACCENTURE")
plt.legend() # to show the label in graph employee data
plt.show()
```

```python
#bar graph with id and salary of 2 department in 2 dataset
x=[1001,1002,1003,1004,1005,1006]   #emp id and sal of sales department
y=[10000,12000,13000,14000,15000,16000]

x1=[1002,1004,1006,1008,1010,1012]
y1=[5000,6000,4500.00,12000,9000,10000]  ## emp id and sal of production department

plt.bar(x,y,label="Sales dept", color="red")
plt.bar(x1,y1,label="Production dept",color="green")

plt.xlabel("employee id")
plt.ylabel("salary")
plt.legend()
plt.show()
```

**Histogram**

```python
## distribution of values useful to show values grouped in bins or intervals. range (0-10,10-20,20-30)
emp_age=[22,45,30,67,56,78,70,34,56,42,76,66,65,20,25,35,80]
bins=[0,10,20,30,40,50,60]
```

```python
plt.hist(emp_age,bins,histtype="bar",rwidth=0.8, color="Blue")  #histype =bar,barstacked, step, stepfilled  r width=.8 bar width 80%, gap of 10% before and after bar
#set labels
plt.xlabel("employee ages")
plt.ylabel("no of employees")
plt.title("microsoft")

#draw the histogram
plt.show()
```

***Creating PIE chart***

```python
# pie chart shows a circle that is divide into parts each part represents a proportion of the whole. expenditure of 4 department
# take the expenditure of 4 department
slices=[50,14,24,16]
depts=["sales","production","HR","Finance"]

#tak ethe colors for each department
cols=["magenta","cyan","brown","gold"]

#craete pie vhart
plt.pie(slices, labels=depts,colors=cols,startangle=90,explode=(0,0.2,0,0),shadow=True, autopct="%.1f%%")

#set title
plt.title("wipro EXPENDITURE")
plt.legend()

plt.show()
```

**LINE GRAPH**

```python
# a line graph is simple graph that shows the results in the form of lines.nto cfreate line graph v need nseries of points
# to display profit if company
years=["2015","2016","2017","2018","2019","2020","2021"]
profits=[9,10,10.5,8.9,12.75,7.51,6.45]
x=years
y=profits
# craete line graph
plt.plot(x,y,"blue")

#set title and labels
plt.xlabel("years")
plt.ylabel("profits")
plt.title("GOOGLE PROFIT PER YEAR")
plt.legend()
plt.show()
```

**SCATTER PLOT**

```python
# to display relationship between age and price of houses at delhi
import numpy as np

#take age and prices as lists
#age=[6,8,9,2,10,2,9,4,11,12,9] # data as a list
age=np.array([6,8,9,2,10,2,9,4,11,12,9])  # data as 1D array
price=[97,87,82,120,77,122,99,97,74,79,83]

# draw scatter plot
plt.scatter(age,price,color="red")
plt.title("age and price at delhi")
plt.xlabel("age i  years")
plt.ylabel("price i lakhs of rs")
plt.show()
```
